```python
#here's a generic function that will plot a chosen axis
#as well as the hyperplane perpendicular to the chosen axis 
#as well as show some residuals from the perpendicular hyperplane 
#for various data points

def plot_axis(data, slope):
    line = lambda x: slope * x

    vec = np.array([1, line(1)])
    dx = np.dot(data, vec) / (np.linalg.norm(vec)**2)
    dy = vec[1] * dx
    xs = data[:, 0] - dx
    ys = data[:, 1] - dy

    plt.scatter(data[:, 0], data[:, 1])

    plt.plot([-3, 4], [line(-3), line(4)], color='r', linewidth=4)

    if slope:
        line_perp = lambda x: -(1./slope) * x 
        plt.plot([-3, 3], [line_perp(-3), line_perp(3)], color='r', linewidth=2, linestyle='--')
    else:
        plt.plot([0, 0], [-3, 3], color='r', linewidth=2, linestyle='--')
    
    for i in range(1, 41, 5):
        x0, y0 = data[i]
        x1 = xs[i]
        y1 = ys[i]
        plt.plot([x0, x1], [y0, y1], color='y', linewidth=3)

    plt.axis('equal')
    ax = plt.gca()
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
```

```python
#This is with the slope picking out the correct principal axis
plot_axis(data=toy_data_ctr, slope=9./7.)
```

```python
#now we care to understand how much variance is absorbed by 
#a given axis

def total_var(data, slope):
    line = lambda x: slope * x
    vec = np.array([1, line(1)])
    dx = np.dot(data, vec) / (np.linalg.norm(vec)**2)
    dy = vec[1] * dx
    diffs = np.sqrt((dx**2 + dy**2))
    return diffs.var()
```

```python
#The correct choice of principal axis absorbs the most variance
slope = 9./7.
plot_axis(toy_data_ctr, slope)
varval = total_var(toy_data_ctr, slope)
plt.title('Variance: %.3f' % varval)
```

```python
#... while the wrong choice absorbs less
slope = 0
plot_axis(toy_data_ctr, slope)
varval = total_var(toy_data_ctr, slope)
plt.title('Variance: %.3f' % varval)
```

```python
#... in fact picking the choice in the orthogonal hyperplane
#minimizes the absorbed variance

slope = - 1. / (9./7.)
plot_axis(toy_data_ctr, slope)
varval = total_var(toy_data_ctr, slope)
plt.title('Variance: %.3f' % varval)
```

```python
#Method of "deflation" removes all the variance
#along the principal axis and just leaves what's left

slope = 9/7.
line = lambda x: slope * x
vec = np.array([1, line(1)])
proj = np.dot(toy_data_ctr, vec) / (np.linalg.norm(vec)**2)
proj = toy_data_ctr - np.outer(proj, vec)
plt.scatter(proj[:, 0], proj[:, 1])

plt.axis('equal')
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
```

## Eigenvalues Examples

$$M = \begin{bmatrix} 2 & 0 \\ 
                    0 & 0.5
    \end{bmatrix}$$
    
so eigenvalues $\lambda$ and eigenvectors $v$ are solutions of

$$\left (\begin{bmatrix} 2 & 0 \\ 
                    0 & 0.5
    \end{bmatrix} - \lambda \right) \cdot v = 0.$$
    
But just expressing things in terms of the components of $v$ and doing the algebra out gives

$$\begin{bmatrix} 2v[0] \\ .5v[1]\end{bmatrix} = \begin{bmatrix} \lambda v[0] \\ \lambda v[1]\end{bmatrix}$$

These are two equations that have to be simultaneously true.  So we have

$$2v[0] = \lambda v[0] \quad \text{ and } \quad 0.5v[1] = \lambda v[1].$$

Assuming $v[0], v[1] \neq 0$, we have

$$\lambda = 2 \quad \text{ and } \quad \lambda = 0.5$$

Obviously these can't both be true simultaneously, so either $v[0] = 0$ or $v[1] = 0$. Thus

$$\lambda = 2 \text{ for } v[1] = 0 \quad \quad ; \quad \quad \lambda = 0.5 \text{ for } v[0] = 0.$$

In other words, our eigenvalues and eigenvectors are

$$\vec{v}_0 = \begin{bmatrix}1 \\ 0 \end{bmatrix} \quad \text{with associated eigenvalue 2}$$ and
$$\vec{v}_2 = \begin{bmatrix}0 \\ 1\end{bmatrix} \quad \text{with associated eigenvalue 0.5}$$



PRINCIPLE: If the matrix is diagonal, than the eigenvalues are just values on the diagonal, and the eigenvectors are the standard basis.

Now let's do another example:

$$M = \begin{bmatrix} 0 & 1 \\ 
                    -1 & 0
    \end{bmatrix}$$
    
so eigenvalues $\lambda$ and eigenvectors $v$ are solutions of

$$\left (\begin{bmatrix} 0 & 1 \\ 
                    -1 & 0
    \end{bmatrix} - \lambda \right) \cdot v = 0.$$
    
But just expressing things in terms of the components of $v$ and doing the algebra out gives

$$\begin{bmatrix} v[1] \\ -v[0]\end{bmatrix} = \begin{bmatrix} \lambda v[0] \\ \lambda v[1]\end{bmatrix}$$

Again, these are two equations that have to be simultaneously true.  But this means that 

$$\lambda = \frac{v[1]}{v[0]} = - \frac{v[0]}{v[1]}$$ 

which is the same as saying that

$$\left ( \frac{v[1]}{v[0]} \right)^2 = -1.$$

Thus 

$$\lambda = \frac{v[1]}{v[0]}  = i \quad \text{or} \quad \lambda = \frac{v[1]}{v[0]}  = -i$$

and the associated eivenvectors are:

$$\vec{v}_0 = \begin{bmatrix}i \\ -1\end{bmatrix} \quad \text{and} \quad \vec{v}_1 = \begin{bmatrix}i \\ 1\end{bmatrix}.$$


PRINCIPLE: 

Dilation matrices have real eigenvalues, and the eigenvalues corresponds to the dilation factors.  

Rotation matrices have imaginary eigenvalues, and the eigenvalues correspond to the rotation angles.



### The Power Method

```python
#ok let's load our usual data 
Neural_Data_subset = np.load('data_subset.npy')
```

```python
def featurewise_norm(data, fmean=None, fvar=None, tol=1e-5):
    """perform a whitening-like normalization operation on the data, feature-wise
       Assumes data = (K, M) matrix where K = number of stimuli and M = number of features
    """
    if fmean is None:
        fmean = data.mean(0)
    if fvar is None:
        fvar = data.std(0)
    data = data - fmean  #subtract the feature-wise mean of the data
    data = data / np.maximum(fvar, tol)  #divide by the feature-wise std of the data
    return data
```

```python
Neural_Data_subset = featurewise_norm(Neural_Data_subset)
```

```python
cov = np.dot(Neural_Data_subset.T, Neural_Data_subset)
```

```python
#number of "feature dimensions" in the dataset
cov.shape
```

```python
def norm(v): return v / np.linalg.norm(v)

def power_method(A, n=200):
    #A = our matrix --- square of shape (k, k)
    #n = number of times we want to iterate the power method
    
    #initial guess for eigenvector
    k = A.shape[1]  #number of dimensions
    v = norm(np.random.rand(k))

    #iterate
    for i in range(n):
        #calculate v --> Av
        v = np.dot(A, v)
        #normalize
        v = norm(v)
        
    #compute corresponding eigenvalue
    l = np.dot(v.T, np.dot(A, v)) / np.dot(v, v)

    return l, v
    
```

```python
def get_eigenvalues(A, n=200):
    #A is a square matrix 
    k = A.shape[1]
    vals = []
    vecs = []
    for i in range(k):
        l, v = power_method(A, n=n)   #power method
        vals.append(l); vecs.append(v) 
        A = A - l * np.outer(v.T, v)  #deflation
    return np.array(vals), np.array(vecs).T
```

```python
eigenvals, eigenvecs = get_eigenvalues(cov)
```
